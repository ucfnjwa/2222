```python
dataoall=pd.read_csv("1.1to6.5(26135).csv")
dataoall.head()
```

```python
import matplotlib.pyplot as plt

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Italy', 'Champions', 'Europe', 'Others'
sizes = [18, 26, 40, 32]
explode = (0.1, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
x1=dataoall["ChuWin"]
y1=dataoall["ChuDraw"]
y2=dataoall["ChuLose"]
fig, ax2 = plt.subplots()
ax2.plot(x1, y1)
ax2.set_ylim([0,10])
ax2.set(xlabel='ChuWin', ylabel='ChuDraw',
       title='Relationship of odds')
ax2.grid()
plt.show()
```

```python
fig, ax2 = plt.subplots()
ax2.plot(x1, y2)
ax2.set_ylim([0,10])
ax2.set(xlabel='ChuWin', ylabel='ChuLose',
       title='Relationship of odds')
ax2.grid()
plt.show()
```
