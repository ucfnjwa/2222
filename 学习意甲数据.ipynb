```python
import pandas as pd
import numpy as np
import requests
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
data=pd.read_csv("yijiashanchu01.csv")
data.head()
```

```python
#增加赔率变量
data["winchg"]=data["chuwin"]-data["win"]
data["drawchg"]=data["chudraw"]-data["draw"]
data["losechg"]=data["chulose"]-data["lose"]

#增加常数项
data["intercept"]=1.0
data.head()
```

```python
train_cols=data.columns[[8,9,10,15,16,17,18]]
logit=sm.Logit(data["result1,0"], data[train_cols])

result = logit.fit()
```

```python
result.summary()
```

```python
# 构建预测集
# 与训练集相似，一般也是通过 pd.read_csv() 读入
# 在这边为方便，我们将训练集拷贝一份作为预测集（不包括 admin 列）
import copy
combos = copy.deepcopy(data)
# 数据中的列要跟预测时用到的列一致
predict_cols = combos.columns[[8,9,10,15,16,17,18]]
 
# 进行预测，并将预测评分存入 predict 列中
combos['predict'] = result.predict(combos[predict_cols])
```

```python
combos.head()
```

```python
#计算有限度的准确率
r=0
a=0
for index,row in combos.iterrows():
    if row["predict"]>0.9:
        a=a+1
        if row["result1,0"]==1:
            r=r+1
ra=r/a
ra        
```

```python
#计算大于并且正确的概率以及投注额相等的投注方式
r=0
a=0
t=0.8
money=0
for index,row in combos.iterrows():
    if (row["predict"]-row["winp"])>t and row["win"]>2.3:
        a=a+1
        if row["result1,0"]==1:
            r=r+1
            money=money+(row["winp"]-1)
        else:
            money=money-1
            
ra=r/a
print("shenglv he lirun",ra,"+",money) 
```
